```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
import tensorflow as tf
from tensorflow import keras
```

```python
#데이터 가져오기

data = pd.read_csv("D:/Son/3-2/deeplearning/report/dataset/car+evaluation/car.data", header=None)
data
```

```python
#헤더 이름 지정

data.columns=['buying','maint','doors','persons','lug_boot','safety','class']
data
```

```python
#데이터타입 확인

data.info()
```

```python
#결측치 확인

data.isna().sum()
```

```python
#object형으로는 알고리즘에 입력이 불가
#object타입을 라벨인코딩하여 정수로 변환

from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()
data['buying']=label.fit_transform(data['buying'])
data.head()
```

```python
data['maint']=label.fit_transform(data['maint'])
data.head()
```

```python
data['doors']=label.fit_transform(data['doors'])
data.head()
```

```python
data['persons']=label.fit_transform(data['persons'])
data.head()
```

```python
data['lug_boot']=label.fit_transform(data['lug_boot'])
data.head()
```

```python
data['safety']=label.fit_transform(data['safety'])
data.head()
```

```python
data['class']=label.fit_transform(data['class'])
data.head()
```

```python
#데이터타입 확인

data.info()
```

```python
#라벨과 데이터로 나누기

x=data
x=x.drop(columns=['class'],axis=1)
X=x.to_numpy()
X
```

```python
#라벨

y=data['class']
y
```

```python
#라벨을 원핫인코딩

encoder =  LabelEncoder()
y1 = encoder.fit_transform(y)

Y = pd.get_dummies(y).values

Y
```

```python
#test와 train 데이터로 나눔

from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0) 
```

```python
X
```

```python
Y
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD,Adam


model = Sequential()

model.add(Dense(16,input_shape=(6,),activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(4,activation='softmax'))

model.compile(Adam(0.02),'categorical_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
model_history=model.fit(x=X_train, y=y_train, epochs=100, batch_size=32,validation_data= (X_test,y_test))
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

loss =model_history.history['loss']
val_loss =model_history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
acc =model_history.history['accuracy']
val_acc =model_history.history['val_accuracy']
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
#Accuracy of the predicted values
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class))
print(confusion_matrix(y_test_class,y_pred_class))
```
