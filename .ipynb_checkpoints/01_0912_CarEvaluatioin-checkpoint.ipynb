```python
import numpy as np

import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import LabelEncoder
```

```python
# 데이터 가져오기

#헤더가 지정되어있지 않기 때문에 header=none으로 설정
data=pd.read_csv('D:/Son/3-2/deeplearning/report/dataset/car+evaluation/car.data',encoding='cp949', header=None)
```

```python
#데이터 확인

data.shape
data
```

```python
#헤더 이름 지정

data.columns=['buying','maint','doors','persons','lug_boot','safety','class']
```

```python
data
```

```python
#데이터타입 확인

data.info()
```

```python
#결측치 확인

data.isna().sum()
```

```python
#object형으로는 알고리즘에 입력이 불가
#object타입을 라벨인코딩하여 정수로 변환

label=LabelEncoder()
data['buying']=label.fit_transform(data['buying'])
data.head()
```

```python
data['maint']=label.fit_transform(data['maint'])
data.head()
```

```python
data['doors']=label.fit_transform(data['doors'])
data.head()
```

```python
data['persons']=label.fit_transform(data['persons'])
data.head()
```

```python
data['lug_boot']=label.fit_transform(data['lug_boot'])
data.head()
```

```python
data['safety']=label.fit_transform(data['safety'])
data.head()
```

```python
data['class']=label.fit_transform(data['class'])
data.head()
```

```python
#데이터타입 확인

data.info()
```

```python
#라벨과 데이터로 나누기

x=data
x=x.drop(columns=['class'],axis=1)
x
```

```python
#라벨

y=data['class']
y
```

```python
#훈련과 테스트 데이터로 나누기

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# SVM

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC

clf_svm = SVC(random_state=0)
clf_svm.fit(x_train, y_train)

pred_svm = clf_svm.predict(x_test)

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

```python
# LR

from sklearn.linear_model import LogisticRegression

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(x_train, y_train)

pred_lr = clf_lr.predict(x_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
# DT

from sklearn.tree import DecisionTreeClassifier

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(x_train, y_train)

pred_dt = clf_dt.predict(x_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
# RT

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

print ("\n--- Random Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(x_train, y_train)
pred = rf_clf.predict(x_test)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```
