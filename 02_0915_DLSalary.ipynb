```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
import tensorflow as tf
from tensorflow import keras

data = pd.read_csv("D:/Son/3-2/deeplearning/report/dataset/Salary_Data/Salary_Data.csv")
data
```

```python
#결측치 확인

data.isna().sum()
```

```python
#라벨은 연봉, 데이터는 연차

y=data.Salary.values
x_df=data.drop('Salary',axis=1).values

x_df.dtype, y.dtype
```

```python
# train/test

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(x_df,y,test_size=0.2,random_state=423)
```

```python
#표준화 작업

# 평균
mean = np.mean(X_train, axis = 0)

# 표준편차
std = np.std(X_train, axis = 0) # 표준편차값

# 표준화값
X_train2 = (X_train - mean) / std
X_test2 = (X_test - mean) / std
```

```python
#모델 설계

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(16, activation = 'relu', input_shape = (1,)))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(1)) 
model.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mse'])

model.summary()
```

```python
#표준화 전 데이터로 훈련시작

N_EPOCHS = 100
history = model.fit(X_train, y_train, epochs = N_EPOCHS, validation_data=(X_test, y_test))
```

```python
model.evaluate(X_test, y_test)
```

```python
# loss 체크
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.plot(range(N_EPOCHS), history.history['loss'], label='train loss')
plt.plot(range(N_EPOCHS), history.history['val_loss'], label='test loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
#표준화 후 데이터로 훈련

model2 = Sequential()
model2.add(Dense(16, activation = 'relu', input_shape = (1,)))
model2.add(Dense(16, activation = 'relu'))
model2.add(Dense(1)) 
model2.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mse'])

model2.summary()
```

```python
#표준화 후 데이터로 훈련

N_EPOCHS = 100
history2 = model2.fit(X_train2, y_train, epochs = N_EPOCHS, validation_data=(X_test2, y_test))
```

```python
model2.evaluate(X_test2, y_test)
```

```python
# loss 체크
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.plot(range(N_EPOCHS), history2.history['loss'], label='train loss')
plt.plot(range(N_EPOCHS), history2.history['val_loss'], label='test loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
