```python
from sklearn.preprocessing import LabelEncoder

label_encode = LabelEncoder()
data['activityID'] = label_encode.fit_transform(data['activityID'])
data['activityID']
```

```python
#Data to Hand

from sklearn.preprocessing import StandardScaler

x_hand = data[['handAcc16_1' , 'handAcc16_2' , 'handAcc16_3' ,
    'handAcc6_1' , 'handAcc6_2' , 'handAcc6_3' ,
    'handGyro1' , 'handGyro2',   'handGyro3' ,      
    'handMagne1',  'handMagne2' , 'handMagne3']]
y_hand= data['activityID']

scaler = StandardScaler()
x_hand = scaler.fit_transform(x_hand)

scaled_x_hand = pd.DataFrame(data = x_hand, columns = ['handAcc16_1' , 'handAcc16_2' , 'handAcc16_3' ,
    'handAcc6_1' , 'handAcc6_2' , 'handAcc6_3' ,
    'handGyro1' , 'handGyro2',   'handGyro3' ,      
    'handMagne1',  'handMagne2' , 'handMagne3'])
scaled_x_hand['label'] = y_hand.values
```

```python
#Data to Chest

from sklearn.preprocessing import StandardScaler

x_chest = data[['chestAcc16_1', 'chestAcc16_2',  'chestAcc16_3',   
    'chestAcc6_1',  'chestAcc6_2',  'chestAcc6_3',
    'chestGyro1',   'chestGyro2',   'chestGyro3',    
    'chestMagne1', 'chestMagne2' , 'chestMagne3']]
y_chest = data['activityID']

scaler = StandardScaler()
x_chest = scaler.fit_transform(x_chest)

scaled_x_chest = pd.DataFrame(data = x_chest, columns = ['chestAcc16_1', 'chestAcc16_2',  'chestAcc16_3',   
    'chestAcc6_1',  'chestAcc6_2',  'chestAcc6_3',
    'chestGyro1',   'chestGyro2',   'chestGyro3',    
    'chestMagne1', 'chestMagne2' , 'chestMagne3'])
scaled_x_chest['label'] = y_chest.values
```

```python
#Model to Hand

from sklearn.model_selection import train_test_split

x_hand_train,x_hand_test, y_hand_train, y_hand_test = train_test_split(scaled_x_hand, 
                                                    scaled_x_hand["label"],
                                                    test_size = 0.2, 
                                                    shuffle = True, 
                                                    random_state = 21)
```

```python
#TIME_STEPS : The length of the time segment.
#STEP : Determines the amount of overlap between two consecutive time segments.
    
    
TIME_STEPS = 40
STEP = 10
```

```python
from scipy import stats

def create_dataset(X, y, time_steps=1, step=1):
    Xs, ys = [], []
    for i in range(0, len(X) - time_steps, step):
        v = X.iloc[i:(i + time_steps)].values
        labels = y.iloc[i: i + time_steps]
        Xs.append(v)        
        ys.append(stats.mode(labels)[0][0])
    return np.array(Xs), np.array(ys).reshape(-1, 1)


x_hand_train, y_hand_train = create_dataset(x_hand_train[['handAcc16_1' , 'handAcc16_2' , 'handAcc16_3' ,
    'handAcc6_1' , 'handAcc6_2' , 'handAcc6_3' ,
    'handGyro1' , 'handGyro2',   'handGyro3' ,      
    'handMagne1',  'handMagne2' , 'handMagne3']], x_hand_train.label, TIME_STEPS,
                                  STEP)
x_hand_test, y_hand_test = create_dataset(x_hand_test[['handAcc16_1' , 'handAcc16_2' , 'handAcc16_3' ,
    'handAcc6_1' , 'handAcc6_2' , 'handAcc6_3' ,
    'handGyro1' , 'handGyro2',   'handGyro3' ,      
    'handMagne1',  'handMagne2' , 'handMagne3']], x_hand_test.label, TIME_STEPS, STEP)

print(x_hand_train.shape, y_hand_train.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown = "ignore", sparse = False)
enc.fit(y_hand_train)

y_hand_train = enc.transform(y_hand_train)
y_hand_test = enc.transform(y_hand_test)
```

```python
#Model to hand

model_hand = Sequential()

model_hand.add(LSTM(64,input_shape = [n_timesteps_hand, n_features_hand], return_sequences=True))

model_hand.add(LSTM(128))

model_hand.add(Dropout(0.3))

model_hand.add(Dense(64, activation="relu"))

model_hand.add(Dropout(0.5))

model_hand.add(Dense(n_outputs_hand, activation = "softmax"))
```

```python
history_hand = model_hand.fit(x_hand_train, y_hand_train, epochs = 10, batch_size = 32, validation_split = 0.1, shuffle = True)
print(model_hand.summary())
```
